# TASK 4: PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avacado_df
```

```python
avacado_df_prophet = avacado_df[['Date','AveragePrice']]
avacado_df_prophet.head()
```

```python
avacado_df_prophet = avacado_df_prophet.rename(columns =  {'Date':'ds','AveragePrice':'y'})
```

```python
avacado_df_prophet
```

# TASK 5: UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

# TASK 6: DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
m = Prophet()
m.fit(avacado_df_prophet)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
forecast  
```

```python
figure =m.plot(forecast,xlabel = 'Date', ylabel = 'Price')
```

```python
figure2 = m.plot_components(forecast)
```
